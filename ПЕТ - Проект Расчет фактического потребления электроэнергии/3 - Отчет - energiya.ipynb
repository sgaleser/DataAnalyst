```python
# Преобразуем столбец с датой
data['date'] = data['date'].map(lambda x: dt.datetime.strptime(x, '%d.%m.%Y'))
# Выделяем из даты день недели
data['DN'] = data.date.dt.dayofweek
# Преобразуем дни недели
data['DN'] = data['DN'].replace(to_replace=0,   value='Пн')
data['DN'] = data['DN'].replace(to_replace=1,   value='Вт')
data['DN'] = data['DN'].replace(to_replace=2,   value='Ср')
data['DN'] = data['DN'].replace(to_replace=3,   value='Чт')
data['DN'] = data['DN'].replace(to_replace=4,   value='Пт')
data['DN'] = data['DN'].replace(to_replace=5,   value='Сб')
data['DN'] = data['DN'].replace(to_replace=6,   value='Вс')
# Заменяем запятую на точку
data['kVt'] = data['kVt'].str.replace(',', '.')
# Преобразуем тип данных
data['kVt'] = data['kVt'].astype('float64')
# Умножить на коэффициент
data['kVt'] = data['kVt'] * ratio
```

**Расчеты:**

***Расходы по 4 ценовой категории***

```python
rashod_4_kat = round(data['kVt'].sum().round(2) * cena_plan_p_4, 2)
```

```python
print('Цена киловатта по 4 категории:', cena_plan_p_4, 'руб')
print()
print('Сумма потребленных киловатт по 4 ценовой категории:', data['kVt'].sum().round(2), 'кВт')
print()
print('Затраты на электроэнергию по 4 ценовой категории:', rashod_4_kat, 'руб')
```

***Расходы по плановым часам пиковой нагрузки***

```python
rashod_plan_peak = round(round(data_paln_peak['kVt'].mean(), 2) * cena_plan_peak, 2)
```

```python
print('Цена киловатт плановых часов пиковой нагрузки:', cena_plan_peak, 'руб')
print()
print('Среднее количество киловатт потребленных по плановой пиковой нагрузки:',round(data_paln_peak['kVt'].mean(), 2), 'кВт')
print()
print('Затраты на плановые часы пиковой нагрузки:', rashod_plan_peak, 'руб')
```

***Расходы по часам максимального совокупного потребления***

```python
rashod_peak = round(round(peak['kVt'].mean(), 2) * cena_peak, 2)
```

```python
print('Цена киловатт на часы максимального совокупного потребления:', cena_peak, 'руб')
print()
print(
    'Среднее количество киловатт потребленных в часы максимального совокупного потребления:'
    ,round(peak['kVt'].mean(), 2), 'кВт')
print()
print('Затраты на часы максимального совокупного потребления:', rashod_peak, 'руб')
```

**Суммарный расход**

```python
# Итоговые затраты на электроэнергию
itog = round(rashod_4_kat + rashod_plan_peak + rashod_peak, 2)
# НДС 20%
nds = round(itog / 100 * 20, 2)
# Затраты на электроэнергию плюс НДС
itog_s_nds = round(itog + nds, 2)
```

```python
print('Суммарный расход:', itog)
print()
print('НДС 20%:', nds)
print()
print('Расход плюс НДС', itog_s_nds)
```
